```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Path al archivo
# agarrar un archivo csv random de la carpeta /Volumes/SSD_Fermin/TP_ANALITICA/data
import os
import random

# Obtener todos los archivos CSV de la carpeta data
data_dir = "/Volumes/SSD_Fermin/TP_ANALITICA/data"
archivos_csv = []
for root, dirs, files in os.walk(data_dir):
    for file in files:
        if file.endswith('.csv') and not file.startswith('._'):
            archivos_csv.append(os.path.join(root, file))

# Seleccionar un archivo aleatorio
archivo = random.choice(archivos_csv)
print(f"Archivo seleccionado: {archivo}")

# Cargar el CSV
df = pd.read_csv(archivo)
```

```python
# Dimensiones del dataset
print("Shape del DataFrame:")
print(df.shape)
```

```python
# Nombres de las columnas
print("Columnas del DataFrame:")
for i, col in enumerate(df.columns, 1):
    print(f"{i}. {col}")
```

```python
# Valores nulos por columna
print("Valores nulos por columna:")
nulos = df.isnull().sum()
porcentaje_nulos = (nulos / len(df)) * 100
pd.DataFrame({
    'Valores Nulos': nulos,
    'Porcentaje Nulos': porcentaje_nulos
}).sort_values('Valores Nulos', ascending=False)
```

```python
# Uso de memoria
def get_size(df):
    return f"{df.memory_usage(deep=True).sum() / 1024**2:.2f} MB"

print(f"Memoria total utilizada por el DataFrame: {get_size(df)}")

# Memoria por columna
memoria_columnas = df.memory_usage(deep=True)
print("\nMemoria utilizada por columna:")
for col, memoria in memoria_columnas.items():
    print(f"{col}: {memoria/1024**2:.2f} MB")
```
