```python
nuevo_dato = {
    'name': 'KONIGSBACHER CERVEZA ALEMANA DE TRIGO 115ML',
    'description': 'Arroyo Craftsman Timber Ridge 13 Inch Tall 1 Light Outdoor Wall Light Timber Ridge - TRB-9HS-CS-S - Rustic',
    'region': 'Outside of North America',
    'imageUrl': '',
    'brand': 'Stella Artois',
    'specs': [['Size', '115 ml'], ['Countries', 'Germany']],
    'category': 'Beer',
    'categoryPath': ['Food, Beverages & Tobacco',
                     'Beverages',
                     'Alcoholic Beverages',
                     'Beer'],
    'upc': None,
    'ean': '1111118',
    'id_producto': '1111118'
}

# Eliminar la fila con id_producto == '1111118'
df_productos = df_productos[df_productos['id_producto'] != 1111118]

# Crear DataFrame con la nueva fila
nuevo_df = pd.DataFrame([nuevo_dato])

# Asegurar que las columnas coincidan (añadir columnas faltantes si fuera necesario)
for col in df_productos.columns:
    if col not in nuevo_df.columns:
        nuevo_df[col] = None
for col in nuevo_df.columns:
    if col not in df_productos.columns:
        df_productos[col] = None

# Reordenar columnas por consistencia
nuevo_df = nuevo_df[df_productos.columns]

# Agregar la fila al DataFrame original
df_productos = pd.concat([df_productos, nuevo_df], ignore_index=True)
```

```python
df_productos.head()
```

```python
# Convertir strings que parecen listas en listas reales
df_productos['specs_parsed'] = df_productos['specs'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) and x.startswith('[[') else [])

# Convertir cada lista de pares en un diccionario
df_productos['specs_dict'] = df_productos['specs_parsed'].apply(lambda x: dict(x) if isinstance(x, list) else {})

# Expandir en columnas
df_specs_expandido = pd.json_normalize(df_productos['specs_dict'])

# Combinar con el dataframe original
df_final = pd.concat([df_productos, df_specs_expandido], axis=1)
```

```python
df_final.head()
```

```python
df_final['name'].value_counts()
```

```python
import re

# Pasar a lowercase y eliminar símbolos comunes
name_clean = df_final['name'].str.lower().str.replace(r'[^\w\s]', ' ', regex=True)

# Regex más flexible: busca algo como "473ml", "473 ml", "1lt", "6x355", etc.
patron_mejorado = r'(\d+(\.\d+)?\s?(ml|l|lt|cc|cl|cm3|cmq|c)\b)'

# Extraer el primer match
df_final['volume_raw'] = name_clean.str.extract(patron_mejorado, expand=False)[0].str.strip()
```

```python
df_final.head()
```

```python
df_final['volume_raw'].value_counts()
```

```python
def convertir_a_ml(valor):
    if pd.isna(valor):
        return None
    valor = valor.replace(',', '.').replace(' ', '')
    try:
        if valor.endswith(('ml', 'c', 'cm3', 'cmq', 'cc')):
            return float(re.sub(r'[^\d\.]', '', valor))
        elif valor.endswith(('cl',)):
            return float(valor.replace('cl', '')) * 10
        elif valor.endswith(('lt', 'l')):
            return float(re.sub(r'[^\d\.]', '', valor)) * 1000
        else:
            return None
    except:
        return None

df_final['volume_ml'] = df_final['volume_raw'].apply(convertir_a_ml)
```
